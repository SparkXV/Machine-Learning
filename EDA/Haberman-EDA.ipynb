```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("haberman.csv")
print(data.shape)
print(data.columns)
```

```python
data['status'].value_counts()
```

```python
data["nodes"].max()
```

```python
data.plot(kind='scatter',x='age',y='nodes',style='whitegrid')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(data,hue="status",height=4) \
   .map(plt.scatter,"age","nodes") \
   .add_legend()
plt.show()
```

# Pair-Plot

```python
# Pairwise scatter plot: Pair plot
# If no of features are high
# Only possible to view 2D patterns

plt.close()
sns.set_style("whitegrid")
#sns.pairplot(data, hue="status",vars=["year", "nodes"],height=3)
sns.pairplot(data,hue="status",height=3)
plt.show()
```

```python
import numpy as np
status_1 = data.loc[data["status"] == 1]           
status_2 = data.loc[data["status"] == 2]
plt.plot(status_1["nodes"], np.zeros_like(status_1['nodes']), 'o')          # plot(x,y,linestyle or marker)
plt.plot(status_2["nodes"], np.zeros_like(status_2['nodes']), 'o')
plt.show()
```

```python
sns.FacetGrid(data, hue="status", height=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(data, hue="status", height=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(data, hue="status", height=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```

```python
counts, bin_edges = np.histogram(status_1['nodes'], bins=10,density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

```

```python

counts, bin_edges = np.histogram(status_1['nodes'], bins=100, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf);
plt.show();
```

```python
print("Means:")
print(np.mean(status_1["age"]))
#Mean with an outlier.
print(np.mean(np.append(status_1["age"],100)));
print(np.mean(status_2["nodes"]))


print("\nStd-dev:");
print(np.std(status_1["age"]))
print(np.std(status_2["age"]))
```
